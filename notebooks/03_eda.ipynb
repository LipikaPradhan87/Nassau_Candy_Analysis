```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("../data/cleaned/nassau_featured.csv")
```

```python
# Sales Distribution
plt.figure(figsize=(10, 5))

ax = sns.histplot(df['Sales'], bins=30, color='skyblue')

for patch in ax.patches:
    height = patch.get_height()
    if height > 0:
        ax.annotate(
            f'{int(height)}',
            (patch.get_x() + patch.get_width() / 2, height),
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.title("Sales Distribution", fontsize=14, fontweight='bold')
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.savefig(
    "../visuals/sales_distribution.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

```python

plt.figure(figsize=(10, 5))

ax = sns.histplot(df['Gross Profit'], bins=30, color='orange')

for patch in ax.patches:
    height = patch.get_height()
    if height > 0:
        ax.annotate(
            f'{int(height)}',
            (patch.get_x() + patch.get_width() / 2, height),
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.title("Gross Profit", fontsize=14, fontweight='bold')
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.savefig(
    "../visuals/gross_profit.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
division_sales = (
    df.groupby('Division')['Sales']
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(10, 6))

bars = plt.bar(
    division_sales.index,
    division_sales.values,
    edgecolor='black',
    color = 'purple',
    linewidth=1
)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'${height:,.0f}',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

plt.title(
    "Total Sales by Division",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel(
    "Division",
    fontsize=12,
    fontweight='bold'
)

plt.ylabel(
    "Total Sales ($)",
    fontsize=12,
    fontweight='bold'
)

plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()

plt.savefig(
    "../visuals/sales_division.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python

region_sales = (
    df.groupby('Region')['Sales']
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(10, 6))

bars = plt.bar(
    region_sales.index,
    region_sales.values,
    edgecolor='black',
    color = 'green',
    linewidth=1
)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'${height:,.0f}',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

plt.title(
    "Total Sales by Region",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel(
    "Division",
    fontsize=12,
    fontweight='bold'
)

plt.ylabel(
    "Total Sales ($)",
    fontsize=12,
    fontweight='bold'
)

plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()

plt.savefig(
    "../visuals/sales_region.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
