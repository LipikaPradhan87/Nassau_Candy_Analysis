```python
#convert Dates
df['Order Date'] = pd.to_datetime(
    df['Order Date'],
    format='%d-%m-%Y'
)

df['Ship Date'] = pd.to_datetime(
    df['Ship Date'],
    format='%d-%m-%Y'
)
```

```python
df.isnull().sum()
```

```python
#Handle Missing Values
df['Units'] = df['Units'].fillna(
    df['Units'].median()
)
```

```python
df = df[df['Sales'] > 0]
df = df[df['Cost'] > 0]
```

```python
df.to_csv("../data/cleaned/cleaned_nassau.csv",
          index=False
        )
```
