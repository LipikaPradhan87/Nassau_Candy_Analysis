```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("../data/cleaned/nassau_featured.csv")
```

```python
product_summary = df.groupby(
    'Product Name'
).agg({
    'Sales':'sum',
    'Gross Profit': 'sum',
    'Units': 'sum',
    'Gross Margin %': 'mean'
}).reset_index()
```

```python
#Top 10 Products by Margin
product_summary.sort_values(
    'Gross Profit',
    ascending=False
).head(10)
```

```python
#High Sales + Low Margin

product_summary[
    (product_summary['Sales'] >
     product_summary['Sales'].median())
    &
    (product_summary['Gross Margin %'] < 20)
]
```
