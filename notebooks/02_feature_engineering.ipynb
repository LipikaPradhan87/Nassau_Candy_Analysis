```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv(
    "../data/cleaned/cleaned_nassau.csv"
)
```

```python
df['Gross Margin %'] = (
    df['Gross Profit']/ df['Sales']
    ) * 100
print(df['Gross Margin %'])
```

```python
df['Profit Per Unit'] = (
    df['Gross Profit']
    /df['Units']
)
print(df['Profit Per Unit'])
```

```python
#Revenue Contribution
total_sales = df['Sales'].sum()
df['Revenue Contribution %'] = (
    df['Sales']
    /total_sales
) * 100
```

```python
#Profit Contribution
total_profit = df['Gross Profit'].sum()
df['Profit Contribution %'] = (
    df['Gross Profit']
    /total_profit
) * 100
```

```python
df.to_csv("../data/cleaned/nassau_featured.csv",
          index=False
        )
```
